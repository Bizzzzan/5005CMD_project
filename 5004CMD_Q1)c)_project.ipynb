```python
import dask.dataframe as dd
import time
from dask.distributed import Client

#define datatypes for each column
dtype={'County Name': 'object',
       'Number of Trips': 'float64',
       'Number of Trips 1-3': 'float64',
       'Number of Trips 10-25': 'float64',
       'Number of Trips 100-250': 'float64',
       'Number of Trips 25-50': 'float64',
       'Number of Trips 250-500': 'float64',
       'Number of Trips 3-5': 'float64',
       'Number of Trips 5-10': 'float64',
       'Number of Trips 50-100': 'float64',
       'Number of Trips <1': 'float64',
       'Number of Trips >=500': 'float64',
       'Population Not Staying at Home': 'float64',
       'Population Staying at Home': 'float64',
       'State Postal Code': 'object'}

n_processors = [10,20]
n_processors_time = {}

for processor in n_processors:

  #initialize Dask Client
  client = Client(n_workers=processor, memory_limit="4GB")

  #print out information about the Dask client (to check the cluster)
  print(client)

  #start timing
  start = time.time()

  #perform Q1)a)i) - Number of people staying at home per week
  df = dd.read_csv("/content/Trips_by_Distance (1).csv", dtype=dtype)
  cleaned_df = df.dropna()
  unique_weeks = cleaned_df['Week'].nunique().compute()
  print(f"Number of unique weeks: {unique_weeks}")
  population_staying_at_home = cleaned_df.groupby(by='Week')['Population Staying at Home'].mean().compute()
  print(population_staying_at_home)

  #stop the timing
  dask_time = time.time() - start
  n_processors_time[processor] = dask_time
  print(f"\nTime with {processor} processors: {dask_time} seconds\n")
  #shutdown the client to free up memory
  client.shutdown()

print("\n", n_processors_time, "\n")


```
